```python
sentence_as_list = []
for sentence in articles:
    sentence_as_list.append(sentence.lower().split())
```

```python
# import modules & set up logging
import gensim, logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
 
# sentences = [['first', 'sentence'], ['second', 'sentence']]
# train word2vec on the two sentences
model = gensim.models.Word2Vec(sentence_as_list, size=8, window=5, min_count=2, workers=32, max_vocab_size = 20000)
```

```python
model.save('models/gensim_w2v_2')
```

```python
model.accuracy('test.txt')
```

```python
dictionary = {key : model[key] for key in model.wv.vocab}
```
